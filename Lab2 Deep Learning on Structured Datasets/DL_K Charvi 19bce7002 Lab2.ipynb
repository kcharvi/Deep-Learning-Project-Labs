### Imports

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from keras.layers import Dropout
from tensorflow.keras.optimizers import SGD
```

# Question 1: Use the above code of "KerasClassifier" on new dataset

### Loading Dataset

```python
dataset=pd.read_csv('/content/drive/MyDrive/Deep learning LAB/L2D2.csv',dtype='object', header=None)
dataset
```

```python
dataset.isnull().any().any()
```

```python
dataset.replace(['R','M'],['0','1'],inplace=True)
dataset.head()
```

```python
dataset=dataset.apply(pd.to_numeric,errors='coerce').fillna(0)
```

```python
dataset.dtypes
```

```python
X = dataset.iloc[:, 0:60]
Y = dataset.iloc[:, 60]
print(X)
print(Y)
```

### Split to training and testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=7)
```

## MODEL 1

```python
model1 = Sequential()
model1.add(Dense(64, input_dim=60, kernel_initializer='uniform', activation='relu'))
model1.add(Dense(60, kernel_initializer='uniform', activation='relu'))
model1.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
model1.compile(
          loss='binary_crossentropy',
          optimizer='adam',
          metrics=['accuracy']
)
```

```python
model1.fit(X_train,
          Y_train,
          epochs=150,
          batch_size=10)
```

```python
scores = model1.evaluate(X_test, Y_test)
print("%s: %.2f%%" % (model1.metrics_names[1], scores[1]*100))
```

### **The accuracy obtained is 85.51% from a model with three layers**


# Question 2: Check whether accuracy will improve with the larger network

## MODEL 2


```python
model2 = Sequential()
model2.add(Dense(256, input_dim=60, kernel_initializer='random_uniform', activation='relu'))
model2.add(Dense(192, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(128, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(92, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(60, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(172, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(148, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(84, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(48, kernel_initializer='uniform', activation='relu'))
model2.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

model2.compile(
          loss='binary_crossentropy',
          optimizer='adam',
          metrics=['accuracy']
)

model2.summary()
```

```python
model2.fit(X_train,
          Y_train,
          epochs=150,
          batch_size=10)
```

```python
scores = model2.evaluate(X_test, Y_test)
print("%s: %.2f%%" % (model2.metrics_names[1], scores[1]*100))
```

### **The accuracy obtained is 84.06% from a larger network, which is a decrease from the previous.**

> **NOTE: We cannot always confirm that it would decrease in some cases taking different layers the accuracy had increased.**






# Question 3: Check whether accuracy will be improved after Dropout or not

## MODEL 3


```python
model3 = Sequential()

#DROP OUT LAYER
model3.add(Dropout(0.2, input_shape=(60,)))

model3.add(Dense(256, kernel_initializer='random_uniform', activation='relu'))
model3.add(Dense(192, kernel_initializer='uniform', activation='relu'))
model3.add(Dense(128, kernel_initializer='uniform', activation='relu'))
model3.add(Dense(92, kernel_initializer='uniform', activation='relu'))
model3.add(Dense(60, kernel_initializer='uniform', activation='relu'))
model3.add(Dense(172, kernel_initializer='uniform', activation='relu'))
model3.add(Dense(148, kernel_initializer='uniform', activation='relu'))
model3.add(Dense(84, kernel_initializer='uniform', activation='relu'))
model3.add(Dense(48, kernel_initializer='uniform', activation='relu'))
model3.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

model3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model3.fit(X_train,
          Y_train,
          epochs=150,
          batch_size=10)
```

```python
scores = model3.evaluate(X_test, Y_test)
print("%s: %.2f%%" % (model3.metrics_names[1], scores[1]*100))
```

**Accuracy is almost similar to the previous network though not an effective network.**

# Question 4: Check whether the accuracy improved or not with decay learning rate; Use SGD optimizer

#### Decay = Learning Rate/ Number of epochs

## MODEL 4

```python


model4 = Sequential()
model4.add(Dropout(0.2, input_shape=(60,)))
model4.add(Dense(256, kernel_initializer='random_uniform', activation='relu'))
model4.add(Dense(192, kernel_initializer='uniform', activation='relu'))
model4.add(Dense(128, kernel_initializer='uniform', activation='relu'))
model4.add(Dense(92, kernel_initializer='uniform', activation='relu'))
model4.add(Dense(60, kernel_initializer='uniform', activation='relu'))
model4.add(Dense(172, kernel_initializer='uniform', activation='relu'))
model4.add(Dense(148, kernel_initializer='uniform', activation='relu'))
model4.add(Dense(84, kernel_initializer='uniform', activation='relu'))
model4.add(Dense(48, kernel_initializer='uniform', activation='relu'))
model4.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

#APPLYING DECAY LR
sgd = SGD(learning_rate=0.1, momentum=0.9, decay=0.1/150)
model4.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
scores = model4.evaluate(X_test, Y_test)
print("%s: %.2f%%" % (model4.metrics_names[1], scores[1]*100))
```

**Accuracy has decreased drastically**
